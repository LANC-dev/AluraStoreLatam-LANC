```python
# =======================================================
#  Importaci贸n de librer铆as y datos actualizada
# =======================================================
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

print(" Columnas correctamente importadas:")
print(tienda1.columns)


# =======================================================
#  1 FACTURACIN TOTAL POR TIENDA
# Facturaci贸n = Precio * Cantidad de cuotas
# =======================================================
fact1 = (tienda1['Precio'] * tienda1['Cantidad de cuotas']).sum()
fact2 = (tienda2['Precio'] * tienda2['Cantidad de cuotas']).sum()
fact3 = (tienda3['Precio'] * tienda3['Cantidad de cuotas']).sum()
fact4 = (tienda4['Precio'] * tienda4['Cantidad de cuotas']).sum()

facturacion = pd.DataFrame({
    'Tienda': ['Tienda 1','Tienda 2','Tienda 3','Tienda 4'],
    'Facturaci贸n Total': [fact1, fact2, fact3, fact4]
})
print("\n FACTURACIN TOTAL:")
display(facturacion)


# =======================================================
#  2锔 CATEGORAS MS POPULARES POR TIENDA
# =======================================================
cat1 = tienda1.groupby('Categor铆a del Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False)
cat2 = tienda2.groupby('Categor铆a del Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False)
cat3 = tienda3.groupby('Categor铆a del Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False)
cat4 = tienda4.groupby('Categor铆a del Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False)

print("\n CATEGORAS MS VENDIDAS:")
print("\nTIENDA 1:\n", cat1)
print("\nTIENDA 2:\n", cat2)
print("\nTIENDA 3:\n", cat3)
print("\nTIENDA 4:\n", cat4)


# =======================================================
#  3锔 PROMEDIO DE CALIFICACIN
# =======================================================
eval1 = tienda1['Calificaci贸n'].mean()
eval2 = tienda2['Calificaci贸n'].mean()
eval3 = tienda3['Calificaci贸n'].mean()
eval4 = tienda4['Calificaci贸n'].mean()

evaluaciones = pd.DataFrame({
    'Tienda': ['Tienda 1','Tienda 2','Tienda 3','Tienda 4'],
    'Evaluaci贸n Promedio': [eval1, eval2, eval3, eval4]
})
print("\n EVALUACIN PROMEDIO:")
display(evaluaciones)


# =======================================================
#  4锔 PRODUCTOS MS Y MENOS VENDIDOS
# =======================================================
top1 = tienda1.groupby('Producto')['Cantidad de cuotas'].sum().idxmax()
top2 = tienda2.groupby('Producto')['Cantidad de cuotas'].sum().idxmax()
top3 = tienda3.groupby('Producto')['Cantidad de cuotas'].sum().idxmax()
top4 = tienda4.groupby('Producto')['Cantidad de cuotas'].sum().idxmax()

low1 = tienda1.groupby('Producto')['Cantidad de cuotas'].sum().idxmin()
low2 = tienda2.groupby('Producto')['Cantidad de cuotas'].sum().idxmin()
low3 = tienda3.groupby('Producto')['Cantidad de cuotas'].sum().idxmin()
low4 = tienda4.groupby('Producto')['Cantidad de cuotas'].sum().idxmin()

productos = pd.DataFrame({
    'Tienda': ['Tienda 1','Tienda 2','Tienda 3','Tienda 4'],
    'M谩s vendido (Producto)': [top1, top2, top3, top4],
    'Menos vendido (Producto)': [low1, low2, low3, low4]
})
print("\n PRODUCTOS MS Y MENOS VENDIDOS:")
display(productos)


# =======================================================
#  5锔 COSTO PROMEDIO DE ENVO
# =======================================================
envio1 = tienda1['Costo de env铆o'].mean()
envio2 = tienda2['Costo de env铆o'].mean()
envio3 = tienda3['Costo de env铆o'].mean()
envio4 = tienda4['Costo de env铆o'].mean()

envios = pd.DataFrame({
    'Tienda': ['Tienda 1','Tienda 2','Tienda 3','Tienda 4'],
    'Costo Promedio de Env铆o': [envio1, envio2, envio3, envio4]
})
print("\n COSTO PROMEDIO DE ENVO:")
display(envios)


print("\n ANLISIS LISTO ")
```
